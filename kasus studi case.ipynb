````python
from re import X
def nh(x):
  if x <= 100 and x > 80:
    return('A')
  elif x <= 80 and x > 60:
    return('B+')
  elif x <= 60 and x > 40:
    return('C')
  elif x <= 40 and x > 20:
    return('D')
  elif x <=20 and x >= 0:
    return('E')
````

````python
x = df['Nilai'].apply(nh)
````
